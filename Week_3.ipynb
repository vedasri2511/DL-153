Implement an MLP by varying bias, weights, and learning rate, and record observations for different learning rate values. Plot a graph showing the relationship between loss (error) and learning rate.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

# Dataset
X, y = make_classification(n_samples=500, n_features=10, random_state=42)

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

learning_rates = [0.0001, 0.001, 0.01, 0.1]
loss_values = []

# Train with different learning rates
for lr in learning_rates:
    mlp = MLPClassifier(hidden_layer_sizes=(10,),
                        learning_rate_init=lr,
                        max_iter=200,
                        random_state=42)
    mlp.fit(X_train, y_train)
    loss_values.append(mlp.loss_)

# Plot
plt.plot(learning_rates, loss_values, marker='o')
plt.xlabel("Learning Rate")
plt.ylabel("Final Loss")
plt.title("Loss vs Learning Rate")
plt.show()
```

Implement an MLP using the Gradient Descent algorithm, and analyze the convergence behavior and performance of the network.

```python
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt

mlp = MLPClassifier(hidden_layer_sizes=(20,),
                    solver='sgd',
                    learning_rate_init=0.01,
                    max_iter=200,
                    random_state=42)

mlp.fit(X_train, y_train)

# Plot convergence
plt.plot(mlp.loss_curve_)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Gradient Descent Convergence")
plt.show()
```

Implement a Perceptron on IRIS Dataset what are your observations with respect to learning rate using Perceptron class in sklearn. (explore all the parameters of Perceptron class).

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data
y = iris.target

# Feature scaling
scaler = StandardScaler()
X = scaler.fit_transform(X)

learning_rates = [0.0001, 0.001, 0.01, 0.1]

for lr in learning_rates:
    model = Perceptron(eta0=lr, max_iter=1000, random_state=42)
    model.fit(X, y)
    acc = model.score(X, y)
    print("Learning Rate:", lr, " Accuracy:", acc)
```

Implement MLP on any dataset and explore what particular parameters works for it such as learning rate, iterations etc

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

mlp = MLPClassifier(hidden_layer_sizes=(50, 30),
                    learning_rate_init=0.001,
                    max_iter=300,
                    random_state=42)

mlp.fit(X_train, y_train)

print("Training Accuracy:", mlp.score(X_train, y_train))
print("Testing Accuracy:", mlp.score(X_test, y_test))
```

Implement MLP with appropriate number of hidden layers and perceptrons to learn hand written digit dataset(MNIST).

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

digits = load_digits()
X = digits.data
y = digits.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

mlp = MLPClassifier(hidden_layer_sizes=(128, 64),
                    learning_rate_init=0.001,
                    max_iter=20,
                    random_state=42)

mlp.fit(X_train, y_train)

print("MNIST Training Accuracy:", mlp.score(X_train, y_train))
print("MNIST Testing Accuracy:", mlp.score(X_test, y_test))
```
